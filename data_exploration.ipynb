```python
dataset2 = dataset.drop(columns=['entry_id', 'pay_schedule', 'e_signed'])
```

```python
fig = plt.figure(figsize=(15,12))
plt.suptitle('Histogram of numerical Columms', fontsize=10)
for i in range(dataset2.shape[1]):
    plt.subplot(6, 3, i + 1)
    f = plt.gca()
    f.set_title(dataset2.columns.values[i])
    
    vals = np.size(dataset2.iloc[:, i].unique())
    if vals >= 100:
        vals = 100
        
    plt.hist(dataset2.iloc[:, i], bins=vals, color='#3F5D5D')
    
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
```

```python
dataset2.corrwith(dataset.e_signed).plot.bar(
figsize=(15,10), title = "Correlation with response variable E-signed",
fontsize=15, rot=60, grid=True, color='#3F5D7D')
plt.show()
```

```python
## Correlation matrix
sns.set(style='white')
```

```python
corr = dataset2.corr()
```

```python
# Generate a mask fo rthe upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
```

```python
# Set up matplotlib figure
f , ax = plt.subplots(figsize=(20, 15))
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
           square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
